#15. Measure the performance using Evaluation Metrics

```python
# Calculate evaluation metrics
# for Regresstion eqaution we evaluation r2 score to check acuuracy of meodel
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, classification_report
acc=r2_score(y_pred,y_test)
acc
```

```python
# error is E in mse
E=y_test-y_pred
E
```

```python
 #se- squared error SE in mse
se = E*E
se
```

```python
 #mse - mean squared error
mse = np.mean(se)
```

```python
mse
```

```python
 #finding mse using sklearn library
mse2 = mean_squared_error(y_test,y_pred)
mse2
```

```python
 #rmse is squqred root of mse
rmse = np.sqrt(mse)
rmse
```
