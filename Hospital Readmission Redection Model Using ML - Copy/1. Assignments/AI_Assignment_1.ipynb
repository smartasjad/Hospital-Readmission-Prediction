#3.Write a python Program to find out given No is Prime or Not.

```python
num = int(input("Enter a number to check is prime or not:"))

if num == 1:
    print("It is not a prime number")

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("It is not a prime number")
            break
    else:
        print("It is a prime number")
```

#5.Write a Python program to calculate electricity bill (accept no of unit from user) according to following criteria: Unit Price First 100 units no charge Next 100 units Rs 5 per Unit After 200 Units Rs 10 per Unit (For Example if input unit is 350 then total bill amount is 2000)


```python
def calculate_electricity_bill(unit):
    if unit <= 100:
        return 0
    elif unit <= 200:
        return (unit - 100) * 0.5
    else:
        return (unit - 200) * 10 + (100 * 5)

units = int(input("Enter number of units used by user: "))
bill_amount = calculate_electricity_bill(units)
print("Bill Amount is", bill_amount)
```

#7.Write a program to accept numbers from 1-7 and display name of day like- 1 for sunday, 2 for Monday and so on.

```python

 def day_name(number):
    days = {
        1: "Sunday",
        2: "Monday",
        3: "Tuesday",
        4: "Wednesday",
        5: "Thursday",
        6: "Friday",
        7: "Saturday"
    }
    return days.get(number, "Invalid Day")

number = int(input("Enter a number from 1 to 7: "))
day = day_name(number)
print("The name of the day for", number, "is", day)
```

#7.Write a python program to check given word is Palindrome or not

```python
def is_palindrome(word):
  return word == word[::-1]
word = str(input("Enter a word: "))
if is_palindrome(word):
  print("word is palindrome")
else:
  print("word is not palindrome")
```
